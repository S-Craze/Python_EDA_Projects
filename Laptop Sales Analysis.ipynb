# EDA Analysis

```python
plt.title("Rating Given By People to Laptop",c="r")
ax = sns.countplot(x = 'Rating',data = df)
plt.figure(figsize=(15,5))

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

*above ploting tells that 162 peoples rates laptop 4.4*

```python
# Name vs Rating
sales_rate = df.groupby(['Name'], as_index=False)['Rating'].sum().sort_values(by='Rating', ascending=False)
plt.figure(figsize=(20,5))
plt.title("Name Vs Rating",c="brown")
sns.barplot(x = 'Name',y= 'Rating' ,data = sales_rate)
plt.show()

#From this plot we can say that people most like the asus laptop other than rest
```

*From this plot we can say that people most like the asus laptop other than rest*

```python
plt.title("Display",c="r")
ax = sns.countplot(x = 'Display(inch)',data = df)
sns.set(rc={'figure.figsize':(15,5)})


for bars in ax.containers:
    ax.bar_label(bars)
```

*this plot tells that 566 laptops have 15.6 inches screen*

```python
#Display Vs Sales
sales_price = df.groupby(['Display(inch)'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False) 
# Plotting the bar plot 
plt.figure(figsize=(20, 5)) 
barplot = sns.barplot(x='Display(inch)', y='Sales', data=sales_price)
# Annotating each bar with its sales count 
for index, row in sales_price.iterrows():
    barplot.text(index, row.Sales, round(row.Sales, 2), color='black', ha="center")
plt.title('Sales by Display Size',c="r",fontsize=20) 
plt.xlabel('Display Size (inch)') 
plt.ylabel('Sales') 
plt.show()
```

```python
# Ram vs Sales/Unit 
sales_ram = df.groupby(['Ram'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)
plt.figure(figsize=(5,5))
plt.title("Sales by Ram",c='r',fontstyle='italic',fontsize=16)
sns.barplot(x = 'Ram',y= 'Sales' ,data = sales_ram)
plt.show()

#on this below graph we can say that maximum laptop were saled of 8gb ram
```

*on this below graph we can say that maximum laptop were saled of 8gb ram*

```python
sales_gen = df.groupby(['Name'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
plt.title("Sales by Name",c='r',fontstyle='italic',fontsize=20)
sns.barplot(x = 'Name',y= 'Sales' ,data = sales_gen)
plt.show()
```

*From above graphs we can see that most of purchased was Asus*

```python
# Name vs Price
sales_price = df.groupby(['Name'], as_index=False)['Price(Rs)'].sum().sort_values(by='Price(Rs)', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
plt.title("Sales by Price",c='r',fontstyle='italic',fontsize=20)
sns.barplot(x = 'Name',y= 'Price(Rs)' ,data = sales_price)
plt.show()
```

*Asus laptop was most highest priced* 

```python
# Name vs Rating
sales_price = df.groupby(['Name'], as_index=False)['Rating'].sum().sort_values(by='Rating', ascending=False)
plt.title("Name vs Rating",c='r',fontstyle='italic',fontsize=20)
sns.barplot(x = 'Name',y= 'Rating' ,data = sales_price)
```

*From this plot we can say that people most high rated laptop was asus laptop other than rest*

```python
sales_price = df.groupby(['Storage'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

sns.set(rc={'figure.figsize':(30,5)})
plt.title("Sales by Storage",c='r',fontstyle='italic', fontsize=20)
sns.barplot(x = 'Storage',y= 'Sales' ,data = sales_price)
plt.show()
```

*these plot tells that people mostly like to buy 512 gb storage laptops*

```python
# Name vs Reviews
sales_price = df.groupby(['Name'], as_index=False)['Reviews'].sum().sort_values(by='Reviews', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
plt.title("Name vs Reviews",c='r',fontstyle='italic',fontsize=16)
sns.barplot(x = 'Name',y= 'Reviews',data = sales_price)
```

*these plot tells that people most reviewd about lenovo laptops*

# Top 10 Most Sold Laptops

```python
fig1, ax1 = plt.subplots(figsize=(20,7))
df.groupby('Name')['Sales'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
plt.show()
```

```python
#scatterplot according to Sum of sales per name

Label=df['Name'].unique()
ax=df.groupby(['Name']).agg({"Sales":"sum"})
plt.scatter(Label,ax,c="red")
plt.legend(["Total Sum of Sales"])
plt.show()
```
